 <h2>Exploratory Data Analysis</h2>
 Exploratory Data Analysis (EDA) is a method used to analyze and summarize datasets.
 It contains information of all the passengers aboard the RMS Titanic, which unfortunately was shipwrecked. This dataset can be used to predict whether a given passenger survived or not.
 

```python

import pandas as pd
titanic=pd.read_csv("train.csv")
```

```python
# Few conclusions
# Missing values in Age,Cabin,Embarked columns 
# few columns have inappropriate data types
# more than 70% values are missing in cabin columns will have to drop
```

```python
#dropping cabin column 
titanic.drop(columns=['Cabin'],inplace=True)
```

```python
# input missing values for age
# Strategy -mean
titanic['Age'].fillna(titanic['Age'].mean(),inplace=True)
```

```python
titanic['SibSp'].value_counts()
```

```python
titanic['Parch'].value_counts()
```

```python
# changing the Data Type
titanic[ 'Survived'] =titanic[ 'Survived' ].astype ( 'category') 
titanic ['Pclass'] =titanic[ 'Pclass' ].astype ('category')
titanic [' Sex' ]=titanic[ 'Sex' ].astype ('category') 
titanic[ 'Age' ] =titanic[ 'Age' ].astype('int')
titanic['Embarked']=titanic ['Embarked'].astype('category')

```

```python

# Highly skewed data alot of people had cheaper tickets

# Survival with Pclass
sns.countplot(x=titanic['Survived'],hue=titanic['Pclass'])
pd.crosstab(titanic[ 'Pclass'], titanic[ 'Survived']).apply (lambda r: round ((r/r.sum())*100,1), axis=1)
```

```python


# Survival with sex
sns.countplot(x=titanic['Survived'],hue=titanic['Sex'])
pd.crosstab(titanic['Sex'],titanic['Survived']).apply(lambda r: round ((r/r.sum())*100,1), axis=1)
```

```python

#  Age vs Survived  
# violin Displays distribution of data
sns.violinplot(x ="Sex", y ="Age", hue ="Survived",
data = titanic, split = True)
```

```python
# Pclass  vs Survived  
group = titanic.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived, annot = True, fmt ="d")
```
